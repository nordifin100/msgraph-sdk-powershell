```python
import pandas as pd
import numpy as np

# Create a sample DataFrame
data = {
    'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'D', 'C'],
    'Region': ['East', 'West', 'East', 'North', 'South', 'North', 'West', 'East', 'South', 'East'],
    'Sales': [100, 150, 120, 80, 200, 90, 110, 160, 250, 70],
    'Quantity': [10, 15, 12, 8, 20, 9, 11, 16, 25, 7],
    'Date': pd.to_datetime(['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05',
                            '2023-01-06', '2023-01-07', '2023-01-08', '2023-01-09', '2023-01-10'])
}
df = pd.DataFrame(data)

print("Original DataFrame:")
display(df)

# --- Data Manipulation Examples ---

print("\n1. Select specific columns (Product and Sales):")
product_sales = df[['Product', 'Sales']]
display(product_sales)

print("\n2. Filter rows where Sales are greater than 100:")
high_sales = df[df['Sales'] > 100]
display(high_sales)

print("\n3. Add a new column 'Revenue' (Sales * Quantity):")
df['Revenue'] = df['Sales'] * df['Quantity']
display(df.head())

print("\n4. Group by 'Product' and calculate total Sales and average Quantity:")
product_summary = df.groupby('Product').agg(
    Total_Sales=('Sales', 'sum'),
    Average_Quantity=('Quantity', 'mean')
).reset_index()
display(product_summary)

print("\n5. Filter for 'East' region and sort by 'Sales' in descending order:")
east_region_sorted = df[df['Region'] == 'East'].sort_values(by='Sales', ascending=False)
display(east_region_sorted)
```

```python
import moviepy.config
moviepy.config.IMAGEMAGICK_BINARY = '/usr/bin/convert-im6.q16'
from moviepy.editor import VideoFileClip, TextClip, CompositeVideoClip, ColorClip

# Load the existing video
video_clip = VideoFileClip("my_simple_video.mp4")

# Create a new text clip
# The text will appear for the first 3 seconds of the video
text = TextClip("This is new text on the video!", fontsize=50, color='yellow', bg_color='transparent')
text = text.set_duration(3).set_position(('center', 'top'))

# Composite the text clip onto the video clip
final_video_with_text = CompositeVideoClip([video_clip, text])

# Write the new video file
output_path = "my_simple_video_with_text.mp4"
final_video_with_text.write_videofile(output_path, fps=video_clip.fps)

print(f"Modified video saved to {output_path}")
```
